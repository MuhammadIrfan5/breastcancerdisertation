# Preprocessing

```python
from sklearn.preprocessing import FunctionTransformer, StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.impute import SimpleImputer
preprocessing = Pipeline([
    ("impute", SimpleImputer(strategy="mean", add_indicator=True)),
    ("transform", FunctionTransformer(np.sqrt, feature_names_out="one-to-one")),
    ("scaler", StandardScaler())
])
label_quality = LabelEncoder()
df["diagnosis"] = label_quality.fit_transform(df.loc[:,'diagnosis'])
df["diagnosis"] = df["diagnosis"].astype(np.int64)
df['diagnosis'].value_counts()
```

```python
from sklearn.preprocessing import FunctionTransformer, StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.impute import SimpleImputer
```

```python
preprocessing = Pipeline([
    ("impute", SimpleImputer(strategy="mean", add_indicator=True)),
    ("transform", FunctionTransformer(np.sqrt, feature_names_out="one-to-one")),
    ("scaler", StandardScaler())
])
label_quality = LabelEncoder()
df["diagnosis"] = label_quality.fit_transform(df.loc[:,'diagnosis'])
df["diagnosis"] = df["diagnosis"].astype(np.int64)
df['diagnosis'].value_counts()
```

# Splitting into train and test

```python

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, random_state=42, test_size=0.3)
```

# Training Model

```python

x,y = train.drop("diagnosis", axis=1), train["diagnosis"]
from sklearn.linear_model import RidgeClassifier

log_reg = make_pipeline(preprocessing, RidgeClassifier())
from sklearn.model_selection import learning_curve, LearningCurveDisplay

train_sizes, train_scores, valid_scores = learning_curve(
    log_reg , x, y, train_sizes=np.linspace(0.01, 1, 40),
    cv=10, scoring="accuracy"
)
```

```python
r = LearningCurveDisplay(train_sizes=train_sizes,
                         train_scores=train_scores,
                         test_scores=valid_scores,
                         score_name="Accuracy")

r.plot()
import matplotlib.pyplot as plt
plt.title("Accuracy")
```

# Support Vector Machine

```python
from sklearn.svm import SVC

svc = make_pipeline(preprocessing, SVC())
from sklearn.model_selection import learning_curve, LearningCurveDisplay

train_sizes, train_scores, valid_scores = learning_curve(
    svc , x, y, train_sizes=np.linspace(0.01, 1, 40),
    cv=10, scoring="accuracy"
)
```

```python
r = LearningCurveDisplay(train_sizes=train_sizes,
                         train_scores=train_scores,
                         test_scores=valid_scores,
                         score_name="Accuracy")

r.plot()
plt.title("Accuracy")
```

# Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier

d_tree = make_pipeline(preprocessing, DecisionTreeClassifier(min_samples_leaf=5, random_state=42))
train_sizes, train_scores, valid_scores = learning_curve(
    d_tree, x, y, train_sizes=np.linspace(0.01, 1, 40), cv=10, scoring="accuracy"
)
r = LearningCurveDisplay(train_sizes=train_sizes,
                         train_scores=train_scores,
                         test_scores=valid_scores,
                         score_name="Accuracy")

r.plot()
# plt.axis([0,400, 0.85, 0.98])
plt.title("Accuracy")
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score

lin_scores = cross_val_score(log_reg, x, y, cv=10, scoring="accuracy")
svc_scores = cross_val_score(svc, x, y, cv=10, scoring="accuracy")
tree_scores = cross_val_score(d_tree, x, y, cv=10, scoring="accuracy")
print("Logistic Regression average accuracy:", lin_scores.mean())
print("SVC average accuracy:", svc_scores.mean())
print("Decision Tree average accuracy:", tree_scores.mean())
plt.figure(figsize=(10, 5))
plt.plot([1]*10, lin_scores, ".")
plt.plot([2]*10, svc_scores, ".")
plt.plot([3]*10, tree_scores, ".")
plt.boxplot([lin_scores, svc_scores, tree_scores], labels=["Logistic Regression", "SVC", "Decision Tree"])
plt.title("Cross val scores")
plt.show()
```

```python
log_reg.fit(x,y)
from sklearn.metrics import accuracy_score
x_test, y_test = test.drop("diagnosis", axis=1), test["diagnosis"]
from sklearn.metrics import classification_report, ConfusionMatrixDisplay

y_pred_log_reg = log_reg.predict(x_test)
print("Logistic Regression Classification Report:")
print(classification_report(y_test, y_pred_log_reg, target_names=["Benign", "Malignant"]))

accuracy = accuracy_score(y_test, y_pred_log_reg)

print(f'Testing Accuracy: {accuracy * 100:.2f}%')
```

```python
svc.fit(x, y)
y_pred_svc = svc.predict(x_test)
print("SVC Classification Report:")
print(classification_report(y_test, y_pred_svc, target_names=["Benign", "Malignant"]))

accuracy = accuracy_score(y_test, y_pred_svc)

# Output the testing accuracy
print(f'Testing Accuracy: {accuracy * 100:.2f}%')
```

```python
d_tree.fit(x,y)
y_pred_dtree = d_tree.predict(x_test)
print("Decision Tree Classification Report:")
print(classification_report(y_test, y_pred_dtree, target_names=["Benign", "Malignant"]))

accuracy = accuracy_score(y_test, y_pred_dtree)

# Output the testing accuracy
print(f'Testing Accuracy: {accuracy * 100:.2f}%')
```
